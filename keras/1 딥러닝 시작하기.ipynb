```python
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=.3)
```

```python
from keras.layers import Input, Dense
from keras import Sequential

inputs = Input(shape=(2, ))  # 1차원
dense1 = Dense(units=10, activation='relu')
dense2 = Dense(units=10, activation='relu')
outputs = Dense(units=2)

model = Sequential([inputs, dense1, dense2, outputs])
model.compile(optimizer='rmsprop',  loss='mse', metrics=['mse'])
```

```python
model.fit(X_train, y_train, epochs= 600)
pred = model.predict(X_test)
```
