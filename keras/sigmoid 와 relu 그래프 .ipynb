```python
x = np.linspace(-5, 5, 100)
def relu(x):
    if x < 0 :
        y = 0
    else: 
        y = x 
    return y 

plt.plot(x, relu(x) )
```

```python
x = np.linspace(-5, 5, 100)    
y = np.where(x > 0  , x  ,  0)
plt.plot(x, y )
```
