- y= x^2의 그래프를 그려봅시다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 100)
plt.plot(x, x**2)
```

```python
x = np.linspace(-1, 1, 100)
plt.plot(x, np.exp(x))
```

```python
x = np.linspace(-5, 5, 100)
y = 1 /  (1 + np.exp(-x) )
plt.plot(x, y )
```

```python
x = np.linspace(-50, 10, 100)
y = np.maximum(x, 0)
plt.plot(x, y )
```
