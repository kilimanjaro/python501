```python
from keras.datasets import imdb

(X_train, y_train), (X_test, y_test)  = imdb.load_data(num_words=200)
```

```python
word_to_index = imdb.get_word_index()

# 2. мқёлҚұмҠӨ мӢңн”„нҠё л°ҳмҳҒ л°Ҹ мҳҲм•Ҫм–ҙ м¶”к°Җ
# мӢӨм ң лӢЁм–ҙ мқёлҚұмҠӨлҠ” 3м”© л°Җл Ө мһҲмңјлҜҖлЎң кё°мЎҙ мқёлҚұмҠӨм—җ 3мқ„ лҚ”н•ҙмӨҚлӢҲлӢӨ.
index_to_word = {v + 3: k for k, v in word_to_index.items()}

# 0, 1, 2, 3лІҲм—җ н•ҙлӢ№н•ҳлҠ” мҳҲм•Ҫм–ҙ м§Җм •
index_to_word[0] = "<PAD>"   # нҢЁл”© (кёёмқҙ л§һм¶”кё°мҡ©)
index_to_word[1] = "<START>" # л¬ёмһҘ мӢңмһ‘ нҶ нҒ°
index_to_word[2] = "<UNK>"   # мӮ¬м „м—җ м—ҶлҠ” лӢЁм–ҙ (num_words м ңн•ңмңјлЎң мқён•ҙ нғҲлқҪн•ң лӢЁм–ҙ)
index_to_word[3] = "<UNUSED>"

# 3. X_train[0] ліөкө¬н•ҳкё°
# мҲ«мһҗлҘј лӢЁм–ҙлЎң л°”кҫј л’Ө кіөл°ұ(' ')мңјлЎң мқҙм–ҙл¶ҷмһ…лӢҲлӢӨ.
decoded_review = ' '.join([index_to_word.get(i, '?') for i in X_train[0]])

print(decoded_review)
```

```python
from keras.preprocessing.sequence import pad_sequences
X_train = pad_sequences(X_train, maxlen=100)
X_test = pad_sequences(X_test, maxlen=100)
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
import keras 
from keras.layers import Input, Dense, SimpleRNN, Embedding

model = keras.Sequential()
model.add(Input(shape=(100,)))
model.add(Embedding(200, 16))
model.add(SimpleRNN(8))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping

model.fit(X_train, y_train, epochs= 100, batch_size=32, 
          callbacks=[EarlyStopping(patience=3, restore_best_weights=True)] )
```

```python
model.evaluate(X_test, y_test)
```
